# ***1.Determine the top three most common cuisines in the dataset.***

```python
# Count the numbers of each cuisine
cuisine_counts = df['Cuisines'].value_counts()

#  top three cuisines
top_three_cuisines = cuisine_counts.head(3)

# Create a table
table_data = pd.DataFrame({'Cuisine': top_three_cuisines.index,
                          })
print("Top Three Common Cuisines:")
print(table_data)
```

The "Top Three Cuisines"*** bar graph*** is shown in the image.

It displays the quantity of eateries serving three different cuisines: Chinese, North Indian, and North Indian-Chinese.

The most restaurants serve North Indian cuisine; North Indian-Chinese and Chinese cuisines follow.

The popularity of different cuisines based on the quantity of restaurants serving them is represented visually in the graph.




```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_three_cuisines.index, top_three_cuisines.values, color='pink')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.title('Top Three Cuisines')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# ***2. Calculate the percentage of restaurants that serve each of the top cuisines.***

```python
# Calculate percentages
total_restaurants = len(df)
percentages = (top_three_cuisines / total_restaurants) * 100
```

```python
table_data = pd.DataFrame({'Number of Restaurants': top_three_cuisines.values,
                           'Percentage of Total Restaurants': percentages.values})

print("Restaurants Details:")
print(table_data)
```

The graph is called "Restaurant Details" and is a bar graph.

It displays the number and percentage of eateries that provide Chinese, North Indian, and North Indian-Chinese cuisines. North Indian cuisine accounts for 936 restaurants, or 9.80% of the total, and is the most popular.

511 restaurants (5.35% of total) serve North Indian-Chinese cuisine.

354 restaurants (3.71% of total) provide Chinese cuisine.

The popularity of different cuisines based on the quantity of restaurants serving them is represented visually in the graph.


```python
# Your DataFrame with cuisine names
restaurant_details = pd.DataFrame({
    'Cuisine': ['North Indian', 'North Indian, Chinese', 'Chinese'],
    'Number of Restaurants':[936, 511, 354],
    'Percentage of Total Restaurants':[9.800021, 5.350225, 3.706418]
})

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for Number of Restaurants
ax.bar(restaurant_details.index, restaurant_details['Number of Restaurants'], color='lightblue', label='Number of Restaurants')

# Plotting the Percentage of Total Restaurants as a line plot
ax.plot(restaurant_details.index, restaurant_details['Percentage of Total Restaurants'], color='red', marker='o', label='Percentage of Total Restaurants')

# Adding labels and title
ax.set_xlabel('Cuisine')
ax.set_ylabel('Count / Percentage')
ax.set_title('Restaurant Details')

# Adding a legend
ax.legend(loc='upper right')

# Adding values on top of bars
for i, val in enumerate(restaurant_details['Number of Restaurants']):
    ax.text(i, val + 20, f'{val}', ha='center')

# Adding values on top of the percentage line plot
for i, val in enumerate(restaurant_details['Percentage of Total Restaurants']):
    ax.text(i, val, f'{val:.2f}%', ha='center', va='bottom')

plt.xticks(restaurant_details.index, restaurant_details['Cuisine'], rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## *1. Identify the city with the highest number of restaurants in the dataset.*

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Dataset.csv')

# Identify the city with the highest number of restaurants
city_with_most_restaurants = df['City'].value_counts().idxmax()

print("City with the highest number of restaurants:", city_with_most_restaurants)
```

# *2. Calculate the average rating for restaurants in each city.*

```python
# Calculate the average rating for restaurants in each city
average_rating_per_city = df.groupby('City')['Aggregate rating'].mean()

print("\nAverage rating for restaurants in each city:")
print(average_rating_per_city)
```

# *3. Determine the city with the highest average rating.*

```python
# Calculate the average rating for restaurants in each city
average_rating_per_city = df.groupby('City')['Aggregate rating'].mean()

# Determine the city with the highest average rating
city_with_Highest_avg_rating = average_rating_per_city.idxmax()
Highest_avg_rating = average_rating_per_city.max()

print("City with the Highest average rating:", city_with_Highest_avg_rating)
print("Highest average rating:", Highest_avg_rating)
```

# *Task: Price Range Distribution*

# *1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.*

The picture is called "Distribution of Price Ranges Among Restaurants" and it is a bar graph.

The various pricing ranges from 0 to 6 are represented by the x-axis.
From 0 to over 4000, the y-axis shows the total number of restaurants in each price range.

As you can see from the graph, the majority of restaurants are located in the first two pricing ranges. As the price range rises, less restaurants are found.

A visual depiction of restaurant affordability is given by this graph, which shows that most of them fall into lower price ranges.

```python
# Load the dataset
df= pd.read_csv('Dataset.csv')


# Plotting histogram to visualize the distribution of price ranges
plt.figure(figsize=(10, 6))
plt.hist(df['Price range'], bins=range(1, 7), align='left', color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xticks(range(1, 7))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Beautify the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().tick_params(axis='both', which='both', length=0)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)

plt.show()
```

# *2. Calculate the percentage of restaurants in each price range category.*

```python
# Assuming percentage_per_price_Range
percentage_df = pd.DataFrame(df['Price range'].value_counts(normalize=True) * 100)
percentage_df.reset_index(inplace=True)
percentage_df.columns = ['Price range', 'Percentage']
percentage_df['Percentage'] = percentage_df['Percentage'].round(2)

print(percentage_df)
```

# *2. Compare the average ratings of restaurants with and without online delivery.*

3.25 is the average rating of eateries that provide online delivery.

Restaurants that do not provide online delivery have a average rating of 2.47.

The average evaluations of restaurants with and without online delivery differ by a difference of 0.78. This implies that restaurants with online delivery typically receive better reviews.

```python
# Assuming `data` is your DataFrame containing restaurant data
average_rating_with_delivery =   df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_delivery =   df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()

# Formatting the averages to two decimal places
average_rating_with_delivery_formatted = " {:.2f}".format(average_rating_with_delivery)
average_rating_without_delivery_formatted = "  {:.2f}".format(average_rating_without_delivery)

# Print the results in bold
print("    \033[1mAverage rating of restaurants with online delivery:\033[0m", average_rating_with_delivery_formatted)
print("  \033[1mAverage rating of restaurants without online delivery:\033[0m", average_rating_without_delivery_formatted)


# Calculate the difference in average ratings
rating_difference = average_rating_with_delivery -   average_rating_without_delivery
print("-" *   60)
# Print the difference
print("  \033[1mDifference in average ratings between restaurants with and without online delivery:\033[0m", "{:.2f}".format(rating_difference))
```
